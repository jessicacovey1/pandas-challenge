```python
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
school_data = pd.read_csv(school_data_to_load)
school_data.head()
```

```python
student_data_to_load = "Resources/students_complete.csv"
student_data = pd.read_csv(student_data_to_load)
student_data.head()
```

```python
school_student_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
school_student_data_complete.columns
```

```python
#Calculate the total number of schools
#Calculate the total number of students
#Calculate the total budget
#Calculate the average math score 
#Calculate the average reading score
#Calculate the percentage of students with a passing math score (70 or greater)
#Calculate the percentage of students with a passing reading score (70 or greater)
#Calculate the percentage of students who passed math and reading (% Overall Passing)
#Create a dataframe to hold the above results

#Optional: give the displayed data cleaner formatting

school_count = len(school_data["school_name"].unique())
student_count = len(student_data["Student ID"].unique())
total_budget = school_data["budget"].sum()
avg_math_score = student_data["math_score"].mean()
avg_reading_score = student_data["reading_score"].mean()

passing_math_score = len(student_data.loc[student_data["math_score"]>=70])
passing_math = passing_math_score / student_count
passing_math_percent = passing_math * 100

passing_reading_score = len(student_data.loc[student_data["reading_score"]>=70])
passing_reading = passing_reading_score / student_count
passing_reading_percent = passing_reading * 100

passing_math_reading_score = len(student_data.loc[(student_data["math_score"]>=70) & (student_data["reading_score"]>=70)])
                              
passing_overall = passing_math_reading_score / student_count

passing_overall_percent = passing_overall * 100

district_summary = pd.DataFrame ({
    "Total Schools" : [school_count],
    "Total Students" : [student_count],
    "Total Budget" : [ total_budget],
    "Average Math Score" : [avg_math_score],
    "Average Reading Score" : [avg_reading_score],
    "Percent Passing Math" : [passing_math_percent],
    "Percent Passing Reading" : [passing_reading_percent],
    "Percent Overall Passing" : [passing_overall_percent
                                ]})

district_summary
```

```python
#School Name
#School Type
#Total Students
#Total School Budget
#Per Student Budget
#Average Math Score
#Average Reading Score
#% Passing Math (The percentage of students that passed math.)
#% Passing Reading (The percentage of students that passed reading.)
#% Overall Passing (The percentage of students that passed math **and** reading.)

school_groupby = school_student_data_complete.groupby("school_name")
school_means = school_groupby.mean()

school_type = school_data.set_index('school_name')
school_type = school_type['type']

school_total_students = school_means['size']
school_total_budget = school_means['budget']

per_student_budget = school_total_budget / school_total_students

school_avg_math_score = school_means['math_score']
school_avg_reading_score = school_means['reading_score']

school_counts = school_groupby.count()
school_counts_math = school_student_data_complete.loc[school_student_data_complete["math_score"]>=70]
school_counts_reading = school_student_data_complete.loc[school_student_data_complete["reading_score"]>=70]

school_counts_math_groupby = school_counts_math.groupby("school_name").count()["student_name"]
passing_math = school_counts_math_groupby / school_total_students
passing_math_percent = passing_math * 100

school_counts_reading_groupby = school_counts_reading.groupby("school_name").count()["student_name"]
passing_reading = school_counts_reading_groupby / school_total_students
passing_reading_percent = passing_reading * 100

passing_math_reading_score = len(school_student_data_complete.loc[(school_student_data_complete["math_score"]>=70) & (school_student_data_complete["reading_score"]>=70)])                             
passing_overall = passing_math_reading_score / school_total_students
passing_overall_percent = passing_overall * 100


school_summary = pd.DataFrame ({
    #"School Name" : school_groupby,
    "School Type" : school_type,
    "Total Students" : school_total_students,
    "Total School Budget" : school_total_budget,
    "Per Student Budget" : per_student_budget,
    "Average Math Score" : school_avg_math_score,
    "Average Reading Score" : school_avg_reading_score,
    "Percent Passing Math" : passing_math_percent,
    "Percent Passing Reading" : passing_reading_percent,
    "Percent Overall Passing" : passing_overall_percent
                    })

school_summary
```
